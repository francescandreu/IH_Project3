```python
main_df = pd.read_csv('..\data\mainDF2.csv')
main_df = main_df.sort_values(by=['Value'])
chosenOffice = main_df.iloc[0]
location_coordinates = [chosenOffice.lon, chosenOffice.lat]
```

```python
# Get coordinates in needed format:
#       Using FourSquare -> Get office location
#                        -> Get near venues
# Create gdf
# Generate Map
```

```python
def addVenuesToList(results): #response['results']
    new_list = []
    for i in results:
        
        name = i["name"]
        address =  i["location"]["formatted_address"]
        lat = i["geocodes"]["main"]["latitude"]
        lon = i["geocodes"]["main"]["longitude"]
            
        type_ = {"typepoint": 
                            {"type": "Point", 
                            "coordinates": [lat, lon]}}
        
        new_list.append({"name":name, "lat":lat, "lon":lon, "type":type_})
    return new_list
```

```python
def addVenuesToExistingList(results, lst): #response['results']
    for i in results:
        
        name = i["name"]
        address =  i["location"]["formatted_address"]
        lat = i["geocodes"]["main"]["latitude"]
        lon = i["geocodes"]["main"]["longitude"]
            
        type_ = {"typepoint": 
                            {"type": "Point", 
                            "coordinates": [lat, lon]}}
        
        lst.append({"name":name, "lat":lat, "lon":lon, "type":type_})
    return lst
```

```python
def generateMapFromList(venue_list):
    df = pd.DataFrame(venue_list)
    gdf_1 = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df["lon"], df["lat"]))
    return gdf_1
```

```python
## Office Location
response = fs.getVenue('Outspark', location_coordinates)
office = addVenuesToList(response['results'])
```

```python
## Schools locations
response = fs.getNearbyResults(query='Elementary School', location=location_coordinates, limit=10)
venues = addVenuesToList(response['results'])

response = fs.getNearbyResults(query='Middle School', location=location_coordinates, limit=10)
venues = addVenuesToExistingList(response['results'], venues)

response = fs.getNearbyResults(query='High School', location=location_coordinates, limit=10)
venues = addVenuesToExistingList(response['results'], venues)
```

```python
## Nightclubs
response = fs.getNearbyResults(query='Nightclub', location=location_coordinates, limit=10)
nighclubs = addVenuesToList(response['results'])
```

```python
## Starbucks
response = fs.getNearbyResults(query='starbucks', location=location_coordinates, limit=10)
starbucks = addVenuesToList(response['results'])
```

```python
## Transport
response = fs.getNearbyResults(query='airport', location=location_coordinates, limit=10)
transport = addVenuesToList(response['results'])

response = fs.getNearbyResults(query='Train Station', location=location_coordinates, limit=10)
transport = addVenuesToExistingList(response['results'], transport)
```

```python
## Basketball
response = fs.getNearbyResults(query='Basketball Court', location=location_coordinates, limit=10)
basketball = addVenuesToList(response['results'])
```

```python
## Vegan
response = fs.getNearbyResults(query='Vegetarian / Vegan Restaurant', location=location_coordinates, limit=10)
vegan = addVenuesToList(response['results'])
```

```python
## Pet
response = fs.getNearbyResults(query='Pet Service', location=location_coordinates, limit=10)
pet = addVenuesToList(response['results'])
```
